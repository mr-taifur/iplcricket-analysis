```python
#IPL Data Analysis with Pandas, Numpy, Matplotlib, Seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

print("Data Loaded")
print("Matches Shape:", matches.shape)
print("Deliveries Shape:", deliveries.shape)
```

```python
deliveries.head()
```

```python
# Shape of datasets
print("Matches:", matches.shape)
print("Deliveries:", deliveries.shape)

# Missing values
print(matches.isnull().sum())
print(deliveries.isnull().sum())

# Column names
print(matches.columns)
print(deliveries.columns)
```

```python
# Number of matches and deliveries
print("Total matches:", matches.shape[0])
print("Total deliveries:", deliveries.shape[0])

# Columns in datasets
print("Matches columns:", matches.columns)
print("Deliveries columns:", deliveries.columns)

# Check missing values
print(matches.isnull().sum())
print(deliveries.isnull().sum())
```

```python
#Number of Seasons & Matches

print("Seasons:", matches['season'].nunique())
print("Matches:", matches.shape[0])
matches['season'].value_counts().sort_index().plot(kind="bar", figsize=(10,5))
plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()
```

```python
# # Top 10 highest scores by teams in a single match

team_scores = deliveries.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()
top_scores = team_scores.sort_values(by='total_runs', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='total_runs', y='batting_team', data=top_scores, hue='match_id', dodge=False, palette="viridis")
plt.title("Top 10 Highest Team Scores in IPL")
plt.xlabel("Runs")
plt.ylabel("Team")
plt.show()

print(top_scores)
```

```python
#Most Frequent Match Venues
venue_counts = matches['venue'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette="cubehelix")
plt.title("Most Frequent Match Venues")
plt.xlabel("Number of Matches")
plt.ylabel("Venue")
plt.show()
```

```python
# Top 10 partnerships
partnerships = deliveries.groupby(['match_id','batsman','non_striker'])['total_runs'].sum().reset_index()
top_partnerships = partnerships.sort_values(by='total_runs', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='total_runs', y='batsman', data=top_partnerships, palette="plasma")
plt.title("Top 10 Run Partnerships in IPL (by Batsman)")
plt.xlabel("Runs Scored Together")
plt.ylabel("Batsman")
plt.show()

print(top_partnerships)
```

```python
# Most sixes by batsmen
sixes = deliveries[deliveries['batsman_runs'] == 6]

top_six_hitters = sixes['batsman'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_six_hitters.values, y=top_six_hitters.index, palette="rocket")
plt.title("Top 10 Batsmen with Most Sixes in IPL")
plt.xlabel("Number of Sixes")
plt.ylabel("Batsman")
plt.show()

print(top_six_hitters)
```

```python
# Most fours by batsmen
fours = deliveries[deliveries['batsman_runs'] == 4]

top_four_hitters = fours['batsman'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_four_hitters.values, y=top_four_hitters.index, palette="magma")
plt.title("Top 10 Batsmen with Most Fours in IPL")
plt.xlabel("Number of Fours")
plt.ylabel("Batsman")
plt.show()

print(top_four_hitters)
```

```python
#Most Successful Teams

winner_counts = matches['winner'].value_counts().head(10)
sns.barplot(x=winner_counts.values, y=winner_counts.index, palette="viridis")
plt.title("Most Successful Teams in IPL")
plt.xlabel("Wins")
plt.ylabel("Team")
plt.show()
```

```python
#Toss Decision Analysis
sns.countplot(x="toss_decision", data=matches, palette="pastel")
plt.title("Toss Decision Counts")
plt.show()

toss_effect = (matches['toss_winner'] == matches['winner']).mean() * 100
print(f"Winning after winning toss: {toss_effect:.2f}%")
```

```python
#Top 10 Run Scorers##
top_batsmen = deliveries.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="magma")
plt.title("Top 10 Run Scorers in IPL")
plt.xlabel("Runs")
plt.ylabel("Batsman")
plt.show()
```

```python
#Best Bowlers (Wickets)
wickets = deliveries[(deliveries["dismissal_kind"].notnull()) & 
                     (deliveries["dismissal_kind"] != "run out")]

top_bowlers = wickets.groupby("bowler")["dismissal_kind"].count().sort_values(ascending=False).head(10)

sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="coolwarm")
plt.title("Top 10 Wicket-Takers in IPL")
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.show()
```

```python
#Team Performance per Season
season_winners = matches.groupby("season")["winner"].value_counts().unstack().fillna(0)
plt.figure(figsize=(14,7))
season_winners.plot(
    kind="bar",
    stacked=True,
    figsize=(14,7),
    colormap="tab20"
)
plt.title("Team Performance per Season (Number of Wins)", fontsize=14, fontweight="bold")
plt.xlabel("Season", fontsize=12)
plt.ylabel("Number of Wins", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
        
```
